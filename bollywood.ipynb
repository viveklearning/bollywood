```python
movies['story']=movies['story'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['story']=movies['story'].apply(lambda x: [i.replace(" ","") for i in x])
movies['actors']=movies['actors'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
def convert4(obj):
    L=[]
    counter=0
    for i in obj:
        if counter!=4:
            L.append(i)
            counter+=1
        else:
            break
    return L
```

```python
movies['actors']=movies['actors'].apply(convert4)
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['story']+movies['actors']
```

```python
new_df=movies[['imdb_id', 'original_title','poster_path','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['original_title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].original_title)
```

```python
recommend('Soni')
```

```python
recommend('Why Cheat India')
```
